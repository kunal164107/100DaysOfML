```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams['figure.figsize'] = (20.0, 10.0)

data = pd.read_csv('student.csv')
print(data.shape)
data.head()
```

```python
x1 = data['Math'].values
x2 = data['Reading'].values
y = data['Writing'].values

fig = plt.figure()
ax = fig.gca(projection='3d')
ax = Axes3D(fig)
ax.scatter(x1,x2,y,color='k')
ax.set_xlabel('Math', fontsize=30)
ax.set_ylabel('Reading',fontsize=30)
ax.set_zlabel('Writing',fontsize=30)
plt.show()
```

```python
m = len(x1)
x0 = np.ones(m)
X = np.array([x0, x1, x2]).T
theta = np.array([0.0,0.0,0.0])
Y = np.array(y)
alpha = 0.0001
m = len(Y)
# de=np.array([1,2])
```

```python
def cost_function(X,Y,theta): 
    J = np.sum((np.matmul(X,theta) - Y)**2)/(2*m)
    return J
```

```python
initial_cost = cost_function(X,Y,theta)
print(initial_cost)
```

```python
def gradient_descent(X,Y,theta,alpha,iterations):
    
    cost_history = [0]*iterations
    for iteration in range(iterations):
        
        h = np.matmul(X,theta)
      
        h1 = (h-y)*X[:,0]
        h2 = (h-y)*X[:,1]
        h3 = (h-y)*X[:,2]

        h4 = np.sum(h1)
        h5 = np.sum(h2)
        h6 = np.sum(h3)
        
        theta[0] = theta[0] - (alpha/m)*h4
        theta[1] = theta[1] - (alpha/m)*h5
        theta[2] = theta[2] - (alpha/m)*h6 
        
        cost_history[iteration] = cost_function(X,Y,theta)
        
    return theta, cost_history
```

```python
newB, cost_history = gradient_descent(X, Y, theta, alpha, 100000)
print(newB)
print(cost_history[-1])
```

```python
def rmse(y,y_pred):
    rmse = np.sqrt((np.sum((y-y_pred)**2))/m)
    return rmse

def r2_score(y,y_pred):
    mean_y = np.mean(y)
    ssr = np.sum((y-y_pred)**2)
    sst = np.sum((y-mean_y)**2)
    
    return 1-(ssr/sst)
```

```python
Y_pred = X.dot(newB)
rmse = rmse(Y,Y_pred)
r2_score = r2_score(Y,Y_pred)

print(rmse)
print(r2_score)
```

```python
# using Scikit-Learn to build the model

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = np.array([x0,x1,x2]).T
Y = np.array(y)

reg = LinearRegression()
reg = reg.fit(X,Y)
Y_pred = reg.predict(X)

rmse = np.sqrt(mean_squared_error(Y,Y_pred))
r2_score = reg.score(X,Y)

print(rmse)
print(r2_score)
```
